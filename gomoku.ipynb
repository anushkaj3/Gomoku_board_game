### 3.2 Creating the Board

```python
 # Implement code for 3.2 here
"""The function create_board() is used to create the gomoku board using a list data structure,
and set all positions to unoccupied it returns the create board
"""
# 
"""
----Attributes---
size : this value is taken as an input to the function , it stores the size of the board.
board : it is a list of lists to create a board and store the values based on the position 
that the player places a coin, for now we set all positions to empty hence the list is empty.
i : looping variable to set up the rows.
j: loping variable to store the values in individual rows
col : list to store the row values.
"""
def create_board(size):
    #set up a list to create a board and hold empty values for now.
    board=[]
    #loop over the size of the board to get the total number to keep a count of the rows
    for i in range(size):
        col = []
        #loop along the size , this time to save each value in a row  
        for j in range (size):
            #add empty values to each row 
            col.append(" ")
        # append each row to the final list
        board.append(col)
    return board
```

### 3.3 Is the target position occupied?

```python
 # Implement code for 3.3 here
"""The function is_occupied is used to check if any position on the board has 
already been filled it takes the board that was created or updated , and the 
position coordinates x and y as numbers as the input and returns True if the 
position is occupied otherwise it returns false."""

"""Attributes
 board - the board that is updated or created is sent as input.
 x - the row position on the board that needs to be checked.
 y - the column position on the board that needs to be checked.
 """
def is_occupied(board, x, y):
    # check if at the postion x,y on the board if a stone has been placed.
    if board[x][y] == "●" or board[x][y] == "○":
        return True 
    else:
        return False
```

```python
""" 
The function alpha_to_no() is used to create a list of your column headers 
as alphabet and row headers as number this function is used to get the 
position on the board by obtaining the number that would correspond to a column that is 
represented by an alphabte.

"""
"""
Attributes

alpha_no - list to store your row headers and column headers
num - list to store the numeric row headers
alpha - list to store the alphabet column headers
"""

def alpha_to_no(size):
    alpha_no =[]
    num =[]
    alpha=[]
    #we want to store only capital alphabets hence we start with 65
    values = 65
    for i in range(size):
        #this adds the row numbers to the list
        num.append(i)
        
    for j in range(size):
        #chr() gets the alphabet for the unicode 
        alpha.append(chr(values))
        values += 1
    alpha_no.append(num) 
    alpha_no.append(alpha)

    return alpha_no
```

### 3.4 Placing a Stone at a Specific Intersection

```python
""" The function place_on_board is used to place a stone at any position on the board,
it takes the created/updated board , the stone type i.e - "●"  or "○" 
and the position as a tuple of the row and column eg: ('2','F'), as inputs 
and returns true if the stone was placed in the position,otherwise returns 
false if the position was out of range or it was already occupied.

"""

"""
  Attributes
  
  board : the created board.
  stone : the stone value.
  position : tuple of the row and column where the stone is to be placed.
  row_pos : the value of the row.
  alpha_no : this stores the numeric value based on the alphabet from 
             the alpha_to_no function.
  column_pos: holds the numeric column position.
   
"""
    
def place_on_board(board, stone, position):
    #the first value in the tuple is saved as the row position
    row_pos = position[0]
    row_pos = int(row_pos)
    column_position = position[1]
    column_position = column_position.upper()
    #check if row position is greater than the board size
    if row_pos > len(board):
        return False 
    #call alpha_to_no function to get the list of number positions 
    #corresponding to the alphabet
    alpha_no = alpha_to_no(len(board))
    for x in range(len(board)):
        #check if the alphabet sent as input corresponds to any position
        if alpha_no[1][x] == column_position :
            column_pos = int(x)
            break
        else:
            column_pos = " "
        # if no valid column position was obtained return false
    if column_pos == " ":
        
        return False
    #check if the position has already been previously occupied 
    
    if not(is_occupied(board,row_pos,column_pos)):
        board[row_pos][column_pos] = stone
        return True
    else :       
        return False
    

```

### 3.5 Printing the Board

```python
"""
The function print_board is used to print the present gomoku board 
with stones if any positions have been filled.It takes the updated
board - list data structure as an input.
"""
"""
 Attributes
 board - the list of the values on the board
 x,i and j - looping variables to print the desired row and columns
"""
def print_board(board):
    #call alpha_to_no to get the list of alphabets for the columns as per the size of the board
    alpha_no = alpha_to_no(len(board))
    # this for loop is used to print the first row of alphabets
    for x in range(len(board)):
        print(alpha_no[1][x]," ",end ="")
    # go to the next line after printing the alphabets
    print(end="\n")
    
    #This for loop is used to get the row index
    for i in range(len(board)):
        #this for loop is used to get the column index
        for j in range(len(board)):
            #print the value at that paticular position - can be a stone or empty
            print("{}".format(board[i][j]),end="")
            #check if the looping variable j is not equal to the total 
            #size minus 1 in order to print the required number of dashes 
            if(j != (len(board))-1 ):
                print("--",end="")
        print(" ",i,end="\n")
        if (i != (len(board))-1 ):
            for x in range(len(board)):
                print("""|  """,end="")
            
            print(end="\n")
    
```

### 3.6 Check Available Moves

```python
 # Implement code for 3.6 here
"""The function check_available_moves takes the board as an input , and checks if there 
are any more available positions on the board, it returns a tuple of all the avalable positions 
"""

"""
free_position - tuple of the free positions
x, y , pos - looping variables to get the column or row positions

"""
def check_available_moves(board) :
    free_position = []
    #this for loop is used to get the rwo index
    for x in range(len(board)):
        #this for loop is used to get the column index
        for y in range(len(board)):
            # check if the position has already been occupied 
            if (not (is_occupied(board,int(x),int(y)))):
                x = str(x)
                # to get the column position as the alphabet call alpha_to_no
                alpha_no = alpha_to_no(len(board))
                for pos in range(len(board)):
                    #check if the number sent as input corresponds to any position, 
                    #that position is used to find the corresponding alhpabet
                    if alpha_no[0][pos] == y :
                        column_pos = alpha_no[1][pos]
                        # add the row column combination to a list
                        free_position.append((x,column_pos))
    # convert the list to tupple                                        
    
    return free_position

```

### 3.7 Check for the Winner

```python
 # Implement code for 3.7 here
# Test code for 3.7 here [The code in this cell should be commented]

""" 
The function check_winner checks if there is a consecutive set of 5 
of the same coins, it checks the row, columns and the diagonal.
It takes the board as the input and returns the coin if there are 5 consecuitve coins,
it also checks if the board is filled and returns Draw.

"""

"""
 Attributes 
 row - get the row list from the board
 column - get the column list from the board
 row_d and col_d - the row and column index to check diagonals
 diagonal - create a list of the diagonal values
"""
 

def check_winner(board):
        #Check the diagonals of the board for any consecutive set of 5 similar coins 
        for row_d  in range(len(board)):
            for col_d in range(len(board)):               
                
                # To check for the diagonals we select the row and column to 
                #be above 3,becuase any row or column below
                #three will not have 5 positions.
                if row_d > 3:
                    
                    #to get all the diagonals that move upward as the column value increase 
                    if (len(board) - 1 - col_d) > 3:
                        diagonal = []
                        for next_d in range(5):
                            #add the diagonal values to a list to check for 5 consecutive
                            #it gets the diagonal values towards the right
                            diagonal.append(board[row_d - next_d][col_d + next_d])
                            
                        #call the function to check if the required result is found
                        if check_position(diagonal):
                            return check_position(diagonal)
                    
                    #to get all the diagonals that move upward as the column value increase 
                    if col_d > 3:
                        diagonal  = []
                        for next_d in range(5):
                            #add the diagonal values to a list to check for 5 consecutive
                            #it gets the diagonal values towards the left 
                            diagonal.append(board[row_d - next_d][col_d - next_d])
                            
                        #call the function to check if the required result is found
                        if check_position(diagonal):
                            return check_position(diagonal)                            
                    

                #to get all the diagonals that move downwards as the column value decrease 
                if (len(board) - 1 - row_d) > 3:
                    #add the diagonal values to a list to check for 5 consecutive
                    #it gets the diagonal values towards the right
                    if (len(board) - 1 - col_d) > 3:
                        diagonal = []
                        for next_d in range(5):
                            diagonal.append(board[row_d + next_d][col_d + next_d])
                            
                        #call the function to check if the required result is found
                        if check_position(diagonal):
                            return check_position(diagonal)
                    
                    if col_d > 3:
                        diagonal = []
                        for next_d in range(5):
                            #add the diagonal values to a list to check for 5 consecutive
                            #it gets the diagonal values towards the left 
                            diagonal.append(board[row_d + next_d][col_d - next_d])
                            
                        #call the function to check if the required result is found
                        if check_position(diagonal):
                            return check_position(diagonal)
                   
                  
        #this checks all the rows in the board
        for row in board:
            if check_position(row):
                return check_position(row)

        #this checks all the columns in the board
        for i in range(len(board)):
            column_from_row = [[row[i] for row in board]]
        for column in column_from_row :
            if check_position(column):
                return check_position(column)

        #check if there are any more available moves
        if(check_available_moves(board)):
            status = "None"
        else:
            status = "Draw"
        return status
"""

The function check_position is used to check for 5 consecutive 
stones of the same kind. It takes the row or column or diagonal list as input 
and returns the stone for which a consceutive set has been found.


"""

""" 
Attributes 
black_count - keep a track of the consequtive black coins
white_count - keep a track of the consequtive white coins
pointer- keep track of current position and number of positions
value - the list input or row or column or diagonal values
next_position - keep a track of the next position
"""
def check_position(value):
    
    if value:
        # initialize all the counters to zero
        black_count = 0
        white_count = 0
        pointer = 0 
        next_position = value[1]
        #loop through only if the pointer does not exceed the length of the list
        # and the counters of black and white coins are not 5
        while (pointer < len(value)) and (black_count < 5) and (white_count < 5):
            
            if (value[pointer] == "●"):
                if (next_position == "●"):
                    #if a black coin is found add one only if it is in consecutive positions
                    #and it the previous coin was also a black coin
                    if (black_count != 0):
                        black_count += 1 
                    else:
                        black_count =  2
                else:
                    black_count = 0
                
                
                
            elif (value[pointer] == "○"):
                if (next_position == "○"):
                    #if a white coin is found add one only if it is in consecutive positions
                    #and it the previous coin was also a white coin
                    if (white_count!= 0):
                        white_count += 1
                    else :
                        white_count  = 2  
                else:
                    white_count = 0

            pointer += 1
            if (pointer + 1) < len(value) :
                next_position = value[pointer + 1] 
            else :
                next_position = 0
                

        if black_count > 4:
            return "●"

        elif white_count > 4:
            return "○"

        return False
```

### 3.8 Random Computer Player

```python
 # Implement code for 3.8 here
# Test code for 3.8 here [The code in this cell should be commented]
"""
The function random_computer_player plays the part of the computer, , the function 
find all the available valid positions within a 3 * 3 square and randomly select one
of these positions using the random library
"""

"""
Attributes
board - takes the updated board
player_move - a tuple with the row and column at which the player has place their stons
comp_row and comp_col - to get the random row and column coordinate

"""
def random_computer_player(board, player_move):
    #store the row value from the tupple of the players move sent as input
    row = int(player_move[0])
    #store the column value from the tupple of the players move sent as input
    player_col = player_move[1]
    player_col = player_col.upper()
    
    #call alpha_to_no function to get the list of number positions corresponding to the alphabet
    alpha_no = alpha_to_no(len(board))
    for x in range(len(board)):
        #check if the alphabet sent as input corresponds to any position
        if alpha_no[1][x] == player_col :
            column = int(x)
    #get the random values from a 3*3 square matrix around the players position
    #check if the randomly selected position already has a stone placed 
    #then look for another random position
    
    flag = True
    count = 0
    row_col_list =[]

    while(flag == True):
        #check if all eight positions in the 3*3 matrix has been checked
        if (count < 8):
            #compute random values for the column and row within the 3*3 matrix
            comp_row = random.randint(row-1, row+1)
            comp_col = random.randint(column-1, column+1)
            random_rowcol = [comp_row,comp_col]
            #add the position found into a list to make sure the same positions are not checked again
            if random_rowcol not in row_col_list:
                row_col_list.append(random_rowcol)
                count += 1
            if(comp_row >= 0 and comp_row < len(board)):
                if(comp_col >= 0 and comp_col < len(board)):
                    flag = is_occupied(board,comp_row,comp_col)
        #if all 8 positions on the grid are unavailable then pick any position on the board
        else:
            comp_row = random.randint(0, len(board)-1)
            comp_col = random.randint(0, len(board)-1)
            flag = is_occupied(board,comp_row,comp_col)
    
    #convert the column number into the corresponding Alphabet
    alpha_no = alpha_to_no(len(board))
    for pos in range(len(board)):
        #check if the number sent as input corresponds to any position, 
        #that position is used to find the corresponding alhpabet
        if alpha_no[0][pos] == comp_col :
            column_pos = alpha_no[1][pos] 
      
    return (str(comp_row),column_pos)
    
 
```
